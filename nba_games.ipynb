```python
import pandas as pd
import matplotlib.pyplot as plt
from nba_api.stats.static import teams
pd.set_option('display.max_colwidth', 40)
```

```python
nba_teams = teams.get_teams()
```

```python
teams_df = pd.DataFrame(nba_teams)
teams_df.head()
```

```python
atl_id = int(teams_df[teams_df.abbreviation == 'ATL'].id)
```

```python
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.library.parameters import Season
from nba_api.stats.library.parameters import SeasonType

gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=atl_id,
                            season_nullable='2021-22',
                            season_type_nullable=SeasonType.regular)
```

```python
atl_df = pd.DataFrame(gamefinder.get_normalized_dict()['LeagueGameFinderResults'])
atl_df
```

```python
from nba_api.stats.endpoints import playbyplay, winprobabilitypbp

atl_10_pbp = list()
for i in range(10):
    game_id = atl_df.iloc[i].GAME_ID
    df = playbyplay.PlayByPlay(game_id).get_data_frames()[0]
    df2 = winprobabilitypbp.WinProbabilityPBP(game_id).get_data_frames()[0]
    ls = [df, df2]
    atl_10_pbp.append(ls)
```

```python
atl_10_pbp[0][1].dropna(subset=['EVENT_NUM'])
```

```python
atl_10_pbp[0][0]
```

```python
win_df = atl_10_pbp[0][1]
win_df = win_df.dropna(subset=['EVENT_NUM']).set_index('EVENT_NUM')
pbp_df = atl_10_pbp[0][0]
pbp_df['EVENT_NUM'] = pd.to_numeric(pbp_df.EVENTNUM)
pbp_df = pbp_df.dropna(subset=['EVENT_NUM']).set_index('EVENT_NUM')
joined = pbp_df.merge(win_df)
joined2 = joined[['GAME_ID', 'PERIOD', 'HOME_PTS', 'VISITOR_PTS', 'HOMEDESCRIPTION',
                  'NEUTRALDESCRIPTION', 'VISITORDESCRIPTION', 'HOME_PCT', 'VISITOR_PCT']]
```
